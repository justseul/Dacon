```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∞è Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

## üëã Ïã§Ïäµ
---


```python
# Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
train.head()
```

```python
# ÏõêÌï´ Ïù∏ÏΩîÎî© (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# Î™®Îç∏ Ï†ïÏùò

xgb = XGBClassifier()

lgbm = LGBMClassifier()
```

```python
# Î™®Îç∏ ÌïôÏäµ

# X Îäî trainÏóêÏÑú quality Î•º Ï†úÏô∏Ìïú Î™®Îì† Î≥ÄÏàò
# y Îäî trainÏùò qulity Î≥ÄÏàò
X = train_one.drop('quality',axis=1)
y = train['quality']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)


# fit Î©îÏÜåÎìúÎ•º Ïù¥Ïö©Ìï¥ Î™®Îç∏ ÌïôÏäµ(xgb,lgbm)
XGB_model = xgb.fit(X, y)
LGBM_model = lgbm.fit(X, y)
```

```python
# predict Î©îÏÜåÎìúÏôÄ test_one Îç∞Ïù¥ÌÑ∞Î•º Ïù¥Ïö©Ìï¥ ÌíàÏßà ÏòàÏ∏°

xgb_pred = XGB_model.predict(test_one)
lgbm_pred = LGBM_model.predict(test_one)
```

```python
# sample_submission.csv ÌååÏùºÏùÑ Î∂àÎü¨ÏôÄ ÏòàÏ∏°Îêú Í∞íÏúºÎ°ú Ï±ÑÏõå Ï£ºÍ∏∞(xgb,lgbm)
sub = pd.read_csv('//content//data//sample_submission.csv')
sub['quality'] = xgb_pred
sub['quality2'] = lgbm_pred
sub
```

```python
# Ï†ïÎãµÌååÏùº ÎÇ¥Î≥¥ÎÇ¥Í∏∞
sub.to_csv('pred.csv',index=False)
```
