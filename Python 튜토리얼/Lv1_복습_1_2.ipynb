```python
# 판다스 라이브러리를 불러오세요
import pandas as pd
```

## 데이터 불러오기
---

```python
# train.csv 와 test.csv 를 DataFrame 클래스로 불러오세요
train = pd.read_csv('/content/data/train.csv')
test = pd.read_csv('/content/data/test.csv')
```

## 데이터 정보 관찰하기
---

```python
# head() 를 사용해 데이터 정보를 관찰하세요
print(train.head())
print(test.head())
```

## 데이터 행열 갯수 관찰

```python
# shape 를 사용해 데이터 크기를 관찰하세요
print(train.shape)
print(test.shape)
```

## 결측치 확인
---

```python
# info() 를 사용해 결측치가 있는지 확인하세요.
print(train.info())
print('\n')
print(test.info())
```

## 결측치 전처리
---

```python
# dropna() 를 사용해 train 데이터는 결측치를 제거하고
# fillna() 를 사용해 test 데이터 결측치는 0 으로 대체하세요.
# 그리고 결측치의 갯수를 출력하여 확안하세요.
train = train.dropna()
test = test.fillna(0)

print(train.isnull().sum())
print(test.isnull().sum())
```
